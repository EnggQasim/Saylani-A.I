```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1,1,1001)
y = 2 * x+ np.random.randn(*x.shape) * 0.33

#Plot dummy data on graph
plt.scatter(x,y)
```

# Line Function or Our Model for linear Regression

```python
def createLine(x,w=0,b=0):
    return w*x+b #mx+b line equation
```

```python
#print data
plt.scatter(x,y)

#set manual values of weights and bias also called layers parameters
w=2.0087843
b=-0.0057927

#create line with line equation
predict = createLine(x,w,b)


plt.plot(x,predict,c='r')
plt.show()
```

# Linear Regression with Keras 

```python
from keras import models
from keras import layers
```

# Structure your Model or Network

```python
model = models.Sequential()
model.add(layers.Dense(1, activation='relu', input_dim=1))
```

# Compilation configration

```python
model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
```

# Train your Model

```python
history = model.fit(x,y, epochs=100)
```

## Test Model performance on Test Data with evaluate method

```python
model.evaluate(x,y)
```

## Check Summary of our model

```python
model.summary()
```

# Softmax function Example

```python
def softmax(x):
    prob = []
    for i in x:
        b=i/np.sum(x)
        prob.append(b)
    return prob
```

```python
data =np.array([1,3,5,6,7])
softmax(data)
```

```python
np.sum(softmax(data))
```

# Sigmoid Function

```python
def sigmoid(x):
    return 1. / (1. + np.exp(-x))
```

```python
test = np.linspace(-10,10,101)
y1=sigmoid(test)
y1
```

```python
plt.plot(y1)
plt.show()
```

# rectified linear unit (ReLU)

```python
def relu(x):
    x[x<=0] = 0
    return x
```

# Mean Square Error (Loss function, Cost function, objective function)

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
def rmse1(predictions, targets):
    differences = predictions - targets                       #the DIFFERENCEs.
    differences_squared = differences ** 2                    #the SQUAREs of ^
    mean_of_differences_squared = differences_squared.mean()  #the MEAN of ^
    rmse_val = np.sqrt(mean_of_differences_squared)           #ROOT of ^
    return rmse_val  
```

```python
rmse(y,pline)
```

# MNIST Classification problem Solve with Keras

```python
from keras.datasets import mnist

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

## About Train Data Details

```python
print("Train Image Shape: ", train_images.shape)
print("Len of train labels:", len(train_labels))
train_labels
```

## About Test Data Details

```python
print("Train Image Shape: ", test_images.shape)
print("Len of train labels:", len(test_labels))
test_labels
```

## The network architecture

```python
from keras import models
from keras import layers
network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(10, activation='softmax'))
```

## The compilation step

```python
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

## Preparing the image data

```python
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
train_images.size
```

## Preparing the labels

```python
from keras.utils import to_categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
test_labels[0]
```

## Now Train Model (start Training)

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

## Check Model performance of Test Data

```python
 test_loss, test_acc = network.evaluate(test_images, test_labels)
```

```python
print("Accuracy: ",test_acc,"\nTest Loss Value:",test_loss)
```
